```python
# creating a function to get the count of a city number of match
def get_city_match_count(city):
    mask = data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count('Delhi')
```

```python
# geting data for this are match which is play before 2010 and in a particular city
mask1 = data['city']=='Bangalore'
mask2 = data['date']>'2010-01-10'
data[mask1 & mask2].shape[0]
```

```python
# creating a function to geting data for this are match which is play before 2010 and in a particular city
def count_of_match(city,date):
    mask1 = data['city']== city
    mask2 = data['date']>date
    return data[mask1 & mask2].shape[0]
```

```python
count_of_match('Bangalore','2010-01-10')
```
